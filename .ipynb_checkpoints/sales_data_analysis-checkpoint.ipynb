```python
# Importing packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# configures the notebook to display matplotlib plots directly within the notebook rather than in separate windows
%matplotlib inline
```

```python
# Reading the dataset 
sales = pd.read_csv(
    'sales_data.csv',
    parse_dates=['Date']
)
```

1. We should identify and visualize the overall trend in sales over the six-year period. Are sales increasing, decreasing, or staying relatively constant?

```python
# Sorting the DataFrame by 'Year' in ascending order
sales = sales.sort_values(by='Year')
sales
```

Visuals showing the overall trend in sales over the six-years Period

```python
plt.figure(figsize=(15, 5))
plt.plot(sales['Year'], sales['Order_Quantity'], marker='o', linestyle='-',color='green', alpha=0.6)
plt.title('Sales Trend Over Six Years')
plt.xlabel('Year')
plt.ylabel('Sales')
plt.grid(True)
```

Based on the Visual above, we can conclude that sales 

<b>Sales is currently increasing</b>

```python
import seaborn as sns

# Assuming your DataFrame is named 'sales_df'
# Load your data or replace this with your actual DataFrame
# sales_df = pd.read_csv('your_sales_data.csv')

# Assuming 'Date' column is in datetime format, if not, convert it
# sales['Date'] = pd.to_datetime(sales['Date'])

# Set the 'Date' column as the index for time-series analysis
# sales.set_index('Date', inplace=True)

# Resample the data to get total sales for each month
monthly_sales = sales['Order_Quantity'].resample('M').sum()

# Plotting the overall trend
plt.figure(figsize=(12, 6))
sns.lineplot(x=monthly_sales.index, y=monthly_sales.values)
plt.title('Overall Sales Trend Over Six Years')
plt.xlabel('Date')
plt.ylabel('Total Sales (Revenue)')

```

```python
from statsmodels.tsa.seasonal import seasonal_decompose

# Decompose the time series into trend, seasonality, and residual
result = seasonal_decompose(monthly_sales, model='additive')

# Plot the components
plt.figure(figsize=(12, 8))

plt.subplot(4, 1, 1)
plt.plot(result.observed, label='Observed')
plt.legend(loc='upper left')

plt.subplot(4, 1, 2)
plt.plot(result.trend, label='Trend')
plt.legend(loc='upper left')

plt.subplot(4, 1, 3)
plt.plot(result.seasonal, label='Seasonal')
plt.legend(loc='upper left')

plt.subplot(4, 1, 4)
plt.plot(result.resid, label='Residual')
plt.legend(loc='upper left')

plt.tight_layout()
plt.show()
```

2. we should look for seasonal patterns or trends. Are there specific months or seasons with consistently higher or lower sales?

```python
# Function to define various month and their corresponding season
def get_season(month):
    if month in ['December','January','February']:
        return 'Winter'
    elif month in ['March','April','May']:
        return 'Spring'
    elif month in ['June','July','August']:
        return 'Summer'
    else:
        return 'Fall'
```

```python
sales['Season']=sales['Month'].apply(get_season)
```

```python
plt.bar(sales['Season'], sales['Order_Quantity'], color='skyblue')
```

```python
sales['Season'],sales['Order_Quantity'].plot(kind='bar')
```

```python
sales['Month'].value_counts().plot(kind='bar',figsize=(14,6))
```
